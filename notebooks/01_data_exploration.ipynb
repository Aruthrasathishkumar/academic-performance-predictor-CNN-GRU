```python
import os
import pandas as pd
import numpy as np

raw_folder = "../data/raw/"

print("Files in data/raw/:\n")
for f in sorted(os.listdir(raw_folder)):
    size = os.path.getsize(os.path.join(raw_folder, f)) / 1024
    print(f"  {f}  ({size:.1f} KB)")
```

```python
import glob

# This grabs every CSV file in raw/ 
all_files = glob.glob("../data/raw/*.csv")

print(f"Found {len(all_files)} CSV files\n")

# Load each file, add a column showing which file it came from
all_dfs = []

for filepath in sorted(all_files):
    filename = os.path.basename(filepath)
    
    try:
        df = pd.read_csv(filepath)
        df['source_file'] = filename  # track which file each row came from
        all_dfs.append(df)
        print(f"Loaded: {filename}  →  {df.shape[0]} rows, {df.shape[1]} columns")
    except Exception as e:
        print(f"Could not load {filename}: {e}")

print(f"\nTotal dataframes loaded: {len(all_dfs)}")
```

```python
# Check column consistency across all files
print("Columns in each file:\n")
for i, df in enumerate(all_dfs):
    filename = df['source_file'].iloc[0]
    print(f"{filename}:")
    print(f"  {df.columns.tolist()}\n")
```

```python
# Stack all files into one DataFrame
df_all = pd.concat(all_dfs, ignore_index=True)

print("Combined DataFrame:")
print(f"  Shape: {df_all.shape}")
print(f"  Rows: {df_all.shape[0]}")
print(f"  Columns: {df_all.shape[1]}")
print()
print("Column names:")
for col in df_all.columns:
    print(f"  {col}")
print()
df_all.head(3)
```

```python
# The access log file has a very different structure (16574 rows)
# All other files are assignment data (42 rows each)

# Separate them
access_log_dfs = []
assessment_dfs = []

for df in all_dfs:
    filename = df['source_file'].iloc[0]
    if 'users' in filename.lower():
        access_log_dfs.append(df)
        print(f"ACCESS LOG: {filename}  →  {len(df)} rows")
    else:
        assessment_dfs.append(df)

print(f"\nAssignment files: {len(assessment_dfs)}")
print(f"Access log files: {len(access_log_dfs)}")
```

```python
df_access = access_log_dfs[0].copy()

print("Access log shape:", df_access.shape)
print("\nAccess log columns:")
for col in df_access.columns:
    print(f"  {col}")
print()
df_access.head(5)
```

```python
# The columns we want for ML:
# student identifier, assignment name, grade, max grade, time spent
# Let's find the exact names

# Look at the first assignment file's columns
sample = assessment_dfs[0]
print("Sample file:", sample['source_file'].iloc[0])
print("\nAll columns:")
for col in sample.columns:
    print(f"  {col}")
```

```python
# These are the only columns that matter for our ML model
# Adjust names if Cell 7 shows slightly different spellings
CORE_COLS = [
    'Email',
    'Assignment_Name', 
    'Total_Points_Earned',
    'Total_Points_Possible',
    'Time_Spent',
    'source_file'
]

clean_dfs = []
skipped = []

for df in assessment_dfs:
    filename = df['source_file'].iloc[0]
    
    # Find which of our target columns actually exist in this file
    available = [col for col in CORE_COLS if col in df.columns]
    missing = [col for col in CORE_COLS if col not in df.columns]
    
    if missing:
        print(f"WARNING — {filename} is missing: {missing}")
        skipped.append(filename)
        continue
    
    clean_dfs.append(df[available])

print(f"\nSuccessfully extracted: {len(clean_dfs)} files")
print(f"Skipped: {len(skipped)} files")
```

```python
# Stack all clean assignment data into one DataFrame
df_master = pd.concat(clean_dfs, ignore_index=True)

print("Master DataFrame:")
print(f"  Shape: {df_master.shape}")
print(f"  Students: {df_master['Email'].nunique()}")
print(f"  Assignments: {df_master['Assignment_Name'].nunique()}")
print()

# Show missing values
print("Missing values:")
print(df_master.isnull().sum())
print()

df_master.head(10)
```

```python
# Right now we have points earned out of points possible
# We want a single grade percentage (0-100) per student per assignment
# This normalizes across assignments that have different max points

df_master['grade_pct'] = (
    df_master['Total_Points_Earned'] / df_master['Total_Points_Possible'] * 100
).round(2)

# Handle divide by zero (assignments with 0 possible points)
df_master['grade_pct'] = df_master['grade_pct'].fillna(0)

# Time spent — convert from seconds to minutes for readability
df_master['time_minutes'] = (df_master['Time_Spent'] / 60).round(2)

print("Sample of master data:")
print(df_master[['Email', 'Assignment_Name', 
                  'Total_Points_Earned', 'Total_Points_Possible',
                  'grade_pct', 'time_minutes']].head(10))
print()
print("Grade stats:")
print(df_master['grade_pct'].describe())
```

```python
# Pick the first student and show their entire course trajectory
first_student = df_master['Email'].iloc[0]
journey = df_master[df_master['Email'] == first_student].copy()
journey = journey.sort_values('Assignment_Name')

print(f"Student: {first_student}")
print(f"Assignments in data: {len(journey)}\n")

print(journey[['Assignment_Name', 'grade_pct', 
               'time_minutes']].to_string(index=False))
```

```python
# Save to processed folder
df_master.to_csv("../data/processed/assessment_clean.csv", index=False)
df_access.to_csv("../data/processed/access_log_clean.csv", index=False)

print("Saved:")
print(f"  assessment_clean.csv  →  {df_master.shape}")
print(f"  access_log_clean.csv  →  {df_access.shape}")
```

```python
# ============================================================
# FINAL CLEANUP — run this cell to fix all remaining issues
# ============================================================

df = pd.read_csv("../data/processed/assessment_clean.csv")
print(f"Loaded: {df.shape}")

# Step 1: Remove CodeSandbox
df = df[df['Assignment_Name'] != 'CodeSandbox'].copy()
print(f"After removing CodeSandbox: {df.shape}")

# Step 2: Create never_started flag BEFORE filling NaN
# This preserves the information that the student skipped the assignment
df['never_started'] = df['Total_Points_Earned'].isna().astype(int)

# Step 3: Fill remaining NaN with 0
df['Total_Points_Earned'] = df['Total_Points_Earned'].fillna(0)
df['Time_Spent']          = df['Time_Spent'].fillna(0)
df['grade_pct']           = df['grade_pct'].fillna(0)
df['time_minutes']        = df['time_minutes'].fillna(0)

print(f"\nNaN remaining: {df.isnull().sum().sum()}")
print(f"Never-started rows: {df['never_started'].sum()}")

# Step 4: Define correct COURSE ORDER of assignments
# This maps each assignment to its actual position in the curriculum
# Based on the course structure: Labs 0-5, then Assignments, then Labs 7-12
COURSE_ORDER = {
    'IT207-Lab0-ProgrammingReview':                                          0,
    'IT207-Lab1-TheBigPicture':                                              1,
    'IT207-Lab2-SynchronousFileAPIs':                                        2,
    'IT207-Lab3-AsynchronousFilesystemAPI':                                  3,
    'IT207-Lab4-Events':                                                     4,
    'IT207-Lab5-ServerFundamentals':                                         5,
    'IT207-Assignment1- Displaying the User Agent':                          6,
    'IT207-Assignment2- Switching to Async APIs':                            7,
    'IT207-Assignment3 - TODOServer':                                        8,
    'IT207-Lab7-URL_QueryString':                                            9,
    'IT207-Assignment4 - Using the WHATWG API':                             10,
    'IT207 - Lab8 - RESTfulWebService':                                     11,
    'IT207 - Lab8 - RESTfulWebService_New':                                 12,
    'IT207-Assignment5- Chilly Delights REST API Server':                   13,
    'IT207-Lab9-SQL_Review':                                                14,
    'IT207-Lab10-AccessingMySQLServerFromNodejs':                           15,
    'IT207-Lab11-MySQL Stored Procedures':                                  16,
    'IT207-Lab12-Converting the NodejsTerminal App into a REST API':        17,
    'IT207-Assignment6 - BorrowEase: A Node.js & MySQL Library Solution':   18,
    'Midterm Mock Exam Review':                                             19,
}

# Add the sequence number column
df['assignment_order'] = df['Assignment_Name'].map(COURSE_ORDER)

# Check if any assignment didn't get mapped (name mismatch)
unmapped = df[df['assignment_order'].isna()]['Assignment_Name'].unique()
if len(unmapped) > 0:
    print(f"\nWARNING — these assignments were NOT mapped (check spelling):")
    for name in unmapped:
        print(f"  '{name}'")
else:
    print(f"\nAll assignments mapped to course order successfully.")

print(f"\nFinal shape: {df.shape}")
print(f"Assignments: {df['Assignment_Name'].nunique()}")
print(f"Students: {df['Email'].nunique()}")
```

```python
# Print the assignments in COURSE ORDER to verify it makes sense
order_check = df[['Assignment_Name', 'assignment_order']].drop_duplicates()
order_check = order_check.sort_values('assignment_order')

print("Assignments in course order:\n")
for _, row in order_check.iterrows():
    print(f"  {int(row['assignment_order']):2d}. {row['Assignment_Name']}")
```
